```python
import re
```

```python
def wordopt(text):
  # Convert to lowercase
  text = text.lower()

  #Remove URLs
  text = re.sub(r'https?://\S+ | www\.\S+','',text)

  # Remove HTML tags
  text = re.sub(r'<,*?>', '',text)

  # Remove Punctuation
  text = re.sub(r'[^\w\s]', '',text)

  # Remove Digits
  text = re.sub(r'\d', '', text)

  # Remove newline characters
  text = re.sub(r'\n', ' ', text)

  return text
```

```python
news['text'] = news['text'].apply(wordopt)
```
